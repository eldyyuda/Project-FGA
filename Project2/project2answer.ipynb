```python
from datetime import datetime
```

```python
class BankAccount:

    __MIN_BALANCE = 50000

# Pertama buatlah konstruktor dari class Bank berupa 
# 1. nama pemilik akun bank
# 2. kode rekening bank
# 3. tanggal transaksi terakhir
# 4. saldo terakhir

    def __init__(self, owner, account_number):
        self.owner = owner
        self.account_number  = account_number
        self.date="" 
        self.balance= 50000


# Kedua buatlah method untuk menambah uang ke dalam akun pemilik.
    def deposit(self, amount):
        self.balance += amount
        self.date=datetime.today().strftime('%A, %Y-%m-%d %H:%M:%S')
        print("\n Amount Deposited:",amount)
        

# Ketiga buatlah method untuk melakukan penarikan uang dari akun pemilik.
# Apabila pemilik ingin menarik uang lebih dari nilai saldo + minimal saldo maka transaksi gagal dan saldo tidak berubah.
    def withdraw(self, amount):
        if self.balance>=amount:
            self.balance-=amount
            self.date=datetime.today().strftime('%A, %Y-%m-%d %H:%M:%S')
            print("\n Your Withdraw:", amount)
        else:
            print("\n Insufficient balance!!! ")

# Keempat buatlah fungsi __str__ untuk mencetak informasi tentang objek pemilik akun
    def __str__(self):
        print("Account Owner: " , self.owner)
        print("Account Number: " , self.account_number)
        print('Last Transaction:',self.date)
        print("Current balance: " , self.balance)
        
```

```python
# amount = float(input("Masukkan
if __name__ == "__main__":
    try:
        name=input('Insert Your Name:')
        accountNum=int(input('Insert Your Account Number:'))
        action=BankAccount(name,accountNum)
        while 1:
            choice=int(input('1. Deposit\n2. Withdraw \n3. Information \n4. Exit\n'))
            if choice == 1:
                depo=int(input('Insert your deposit:'))
                action.deposit(depo)
            elif choice == 2:
                withdraw=int(input('Insert your withdraw:'))
                action.withdraw(withdraw)
            elif choice == 3:
                action.__str__()
            elif choice == 4:
                break
    except Exception as e:
        print("Need Numeric Input")
        
```
